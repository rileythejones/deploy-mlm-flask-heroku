```python
import pandas as pd 
from sqlalchemy import create_engine
```

```python
engine = create_engine('sqlite:///db.sqlite3', echo=False)
```

```python
df = pd.read_csv('SpotifyAudioFeaturesApril2019.csv')
```

```python
df
```

```python
df.to_sql('songs', index=False, con=engine)
```
